# Object-Oriented Programming (OOP) in Python

## What is OOP?
Object-Oriented Programming (OOP) is a programming paradigm that organizes software design around **objects** rather than functions or logic.  
Objects represent real-world entities and combine both **data** (attributes) and **behavior** (methods).

## How does OOP work?
- OOP works by defining **classes** as blueprints.
- From these classes, we create **objects** (instances).
- Each object can store its own data and perform actions defined by the class.
- This makes programs more modular, reusable, and easier to maintain.

## Why OOP?
- Provides **modularity**: code is organized into logical units.
- Encourages **reusability**: classes and methods can be reused across projects.
- Improves **maintainability**: changes in one part of the code are less likely to break others.
- Models **real-world problems** more naturally by representing entities as objects.

---

## Main Components of OOP in Python

- **Class** → A blueprint that defines attributes and methods.
- **Object** → An instance created from a class.
- **Attributes** → Variables inside a class that hold data about the object.
- **Methods** → Functions inside a class that define the behavior of the object.
- **Inheritance** → Mechanism to create new classes from existing ones, reusing functionality.
- **Encapsulation** → Bundling data and methods together, with controlled access.
- **Polymorphism** → Same method name used in different classes with different behaviors.
- **Abstraction** → Hiding implementation details and exposing only essential features.

# Understanding `__init__`, `self`, and Other Concepts in OOP (Python)

## `__init__` Method
- Called the **constructor** in Python.
- It runs automatically when a new object of the class is created.
- Purpose: to initialize (set up) the object’s attributes with given values.
- Example in our Blinkit class: it sets up `store_name`, `location`, and `items_available` whenever a new Blinkit object is created.

## `self` Keyword
- Refers to the **current object** being created or used.
- Ensures that attributes and methods belong to that specific object.
- Without `self`, Python would not know which object’s data we are referring to.
- Example: `self.store_name` means “the store_name of this particular Blinkit object.”

## Attributes
- Variables inside the class that hold data about the object.
- Defined using `self` inside `__init__`.
- Example: `self.location` stores the location of the Blinkit store.

## Methods
- Functions inside the class that define the behavior of the object.
- They always take `self` as the first parameter to access the object’s data.
- Example: `display_info()` could show details of the Blinkit store.

## Object Creation
- When we write `blinkit_store = Blinkit("Blinkit Express", "Mumbai", 1200)`, Python:
  1. Calls the `__init__` method.
  2. Passes the values `"Blinkit Express"`, `"Mumbai"`, and `1200`.
  3. Assigns them to the attributes of the new object using `self`.

---

### Summary
- **`__init__`** → Constructor, runs when object is created.  
- **`self`** → Refers to the current object, connects attributes/methods to it.  
- **Attributes** → Data stored inside the object.  
- **Methods** → Actions the object can perform.  
- **Object** → Instance created from the class, with its own unique data.

```python
# Define a Class for Blinkit
class Blinkit:
    # Attributes (Data/Properties)
    def __init__(self, store_name, location, items_available):
        self.store_name = store_name      # Attribute: Store name
        self.location = location          # Attribute: Store location
        self.items_available = items_available  # Attribute: Items in stock
    
    # Methods (Behaviors/Functions)
    def display_info(self):
        pass   # Placeholder for method to show store info
```

```python
# Create an Object (Instance of Blinkit)
blinkit_store = Blinkit(
    store_name="Blinkit Express",
    location="Mumbai",
    items_available=1200
)
```

### <span style="color:Green">Methods</span>
Functions defined inside a class that describe the **behavior** of an object.

### <span style="color:Gray">Encapsulation</span>
The practice of bundling **data (attributes)** and **methods (functions)** together inside a class.  
It also provides **controlled access** to data, often using public, private, or protected visibility.  
Encapsulation ensures that internal details are hidden and only necessary parts are exposed.

---

## <span style="color:Purple">Polymorphism</span>
The ability to use the **same method name** in different classes but with different behaviors.  
It allows flexibility and dynamic behavior in programs.  
Example idea: A `Bird` class and an `Airplane` class may both have a method `fly()`, but each works differently.

# Inheritance in Python OOP

## What is Inheritance?
Inheritance is the mechanism by which a **new class (child/subclass)** can derive properties and behaviors from an **existing class (parent/superclass)**.  
It allows the child class to reuse, extend, or modify the functionality of the parent class.

---

## Where does it work?
- Works in situations where multiple classes share common features.  
- Instead of rewriting the same code, we define it once in a parent class and let child classes inherit it.  
- Example idea: A `Vehicle` class can define general attributes (speed, fuel), while `Car` and `Bike` classes inherit those features and add their own specifics.

---

## Why is it used?
- **Code Reusability** → Avoids duplication by reusing parent class logic.  
- **Extensibility** → Child classes can add new features or override existing ones.  
- **Maintainability** → Changes in the parent class automatically reflect in child classes.  
- **Hierarchy Representation** → Models real-world relationships (e.g., Animal → Dog, Cat).  

---

### Summary
Inheritance helps us:
- Reuse existing code.  
- Extend functionality easily.  
- Represent natural hierarchies in programming.  


```python
# Parent Class (Superclass)
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def display_info(self):
        print("Brand:", self.brand)
        print("Model:", self.model)

# Child Class (Subclass) inheriting from Vehicle
class Car(Vehicle):
    def __init__(self, brand, model, doors):
        # Call the parent class constructor
        super().__init__(brand, model)
        self.doors = doors
    
    # New method specific to Car
    def display_car_info(self):
        self.display_info()  # Reuse parent method
        print("Number of Doors:", self.doors)

# Another Child Class (Subclass) inheriting from Vehicle
class Bike(Vehicle):
    def __init__(self, brand, model, cc):
        # Call the parent class constructor
        super().__init__(brand, model)
        self.cc = cc
    
    # New method specific to Bike
    def display_bike_info(self):
        self.display_info()  # Reuse parent method
        print("Engine Capacity (cc):", self.cc)
```

```python
# Create objects of Child Classes
car1 = Car("Toyota", "Corolla", 4)
```

```python
bike1 = Bike("Honda", "CBR", 650)
```

```python
# Using methods
car1.display_car_info()
```

```python
bike1.display_bike_info()
```

# 🎭 Polymorphism in Python OOP

---

## <span style="color:DarkOrange">What is Polymorphism?</span>
Polymorphism means **"many forms"**.  
It allows the same method name to be used in different classes, but each class can implement it in its own way.  
This makes code more flexible and dynamic.

---

## <span style="color:MediumSeaGreen">Where does it work?</span>
- Works when multiple classes share a **common behavior** but perform it differently.  
- Example idea: Both `Car` and `Bike` classes may have a method called `move()`.  
  - For `Car`, `move()` means driving on four wheels.  
  - For `Bike`, `move()` means riding on two wheels.

---

## <span style="color:RoyalBlue">Why is it used?</span>
- **Flexibility** → Same interface, different implementations.  
- **Code Reusability** → Write general code that works across multiple classes.  
- **Extensibility** → Add new classes without changing existing code.  
- **Real-world modeling** → Different objects behave differently but share a common action.

---

## <span style="color:Purple">Types of Polymorphism</span>
1. **Compile-time Polymorphism (Method Overloading)**  
   - Same method name with different parameters.  
   - Python doesn’t support this directly, but default arguments can mimic it.  

2. **Run-time Polymorphism (Method Overriding)**  
   - Child class redefines a method from the parent class.  
   - The method name stays the same, but behavior changes.  

---

## <span style="color:Crimson">Example Idea (Conceptual)</span>
- **Parent Class → Vehicle**  
  Defines a method `move()`.  
- **Child Class → Car**  
  Overrides `move()` to mean "driving".  
- **Child Class → Bike**  
  Overrides `move()` to mean "riding".  

Even though the method name is the same (`move()`), the behavior differs depending on the object.

---

# Summary
- **Polymorphism** → Same method name, different behaviors.  
- **Where** → Works across multiple classes with shared actions.  
- **Why** → Provides flexibility, reusability, and real-world modeling.  
- **Types** → Method Overloading (compile-time), Method Overriding (run-time).  


```python
# Parent Class (Superclass)
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def move(self):
        print("This vehicle can move.")

# Child Class (Subclass) inheriting from Vehicle
class Car(Vehicle):
    def __init__(self, brand, model, doors):
        super().__init__(brand, model)
        self.doors = doors
    
    # Overriding the move method
    def move(self):
        print(f"The car {self.brand} {self.model} drives smoothly on roads with {self.doors} doors.")

# Another Child Class (Subclass) inheriting from Vehicle
class Bike(Vehicle):
    def __init__(self, brand, model, cc):
        super().__init__(brand, model)
        self.cc = cc
    
    # Overriding the move method
    def move(self):
        print(f"The bike {self.brand} {self.model} rides swiftly with a {self.cc}cc engine.")

# Demonstration of Polymorphism
vehicles = [
    Car("Toyota", "Corolla", 4),
    Bike("Honda", "CBR", 650),
    Vehicle("Generic", "Transport")
]

for v in vehicles:
    v.move()   # Same method name, different behavior
```
