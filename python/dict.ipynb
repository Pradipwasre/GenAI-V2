# Python Dictionary : Deep Dive

## 1. Introduction
A **dictionary** in Python is a collection of **key–value pairs**.  
- Keys act like identifiers.  
- Values are the data associated with those keys.  
- Dictionaries are mutable, dynamic, and allow fast lookups.

---

  # Python Dictionary : Theory

## Definition
A **dictionary** in Python is a built‑in data structure that stores information as **key–value pairs**.  
It is used to represent mappings between unique identifiers (keys) and associated data (values).

---

## Characteristics
- **Keys** must be unique and immutable (e.g., strings, numbers, tuples).  
- **Values** can be of any data type (string, number, list, another dictionary, etc.).  
- Dictionaries are **mutable**, meaning they can be updated after creation.  
- They maintain **insertion order** (from Python 3.7 onwards).  
- Provide **fast lookups** for values using keys.

---

## Operations
- **Accessing**: Retrieve values using their keys.  
- **Adding**: Insert new key–value pairs.  
- **Updating**: Modify values associated with existing keys.  
- **Deleting**: Remove key–value pairs.  
- **Iteration**: Loop through keys, values, or both.  

---

## Methods
- `.get()` → Safely access values without error.  
- `.keys()` → Returns all keys.  
- `.values()` → Returns all values.  
- `.items()` → Returns key–value pairs.  
- `.update()` → Merge another dictionary or update values.  
- `.setdefault()` → Insert a key with a default value if not present.  
- `.pop()` / `.popitem()` → Remove items.  
- `.clear()` → Empty the dictionary.

---

## Advanced Concepts
- **Dictionary Comprehension**: Create dictionaries dynamically using expressions.  
- **Nested Dictionaries**: Store dictionaries inside dictionaries for hierarchical data.  
- **Use Cases**: JSON data representation, configuration storage, frequency counting, mapping IDs to details.

---

## Key Takeaways
- Dictionaries are central to Python programming for structured data.  
- They combine flexibility (mutable values) with efficiency (fast lookups).  
- Ideal for scenarios requiring mappings or associations between data.

## 3. Accessing Values

````python
# Empty dictionary
empty_dict = {}
````

````python
# With initial values
student = {
    "name": "Pradip",
    "age": 28,
    "skills": ["Python", "GenAI", "LangChain"]
}
````

## 3. Accessing Values

````python
print(student["name"])   # Pradip
````

````python
print(student["skills"]) # ['Python', 'GenAI', 'LangChain']
````

````python
# Safe access with .get()
print(student.get("age"))        # 28
print(student.get("location"))   # None (no error)
````

## 4. Adding & Updating

````python
student["location"] = "Pune"   # Add new key–value
student["age"] = 29            # Update existing value
````

## 5. Removing Items

````python
del student["skills"]          # Delete key–value
removed = student.pop("age")   # Remove and return value
student.clear()                # Empty dictionary
````

## 6. Dictionary Methods

````python
# Keys, Values, Items
print(student.keys())    # dict_keys([...])
print(student.values())  # dict_values([...])
print(student.items())   # dict_items([...])
````

````python
# Update with another dict
student.update({"batch": "Gen AI", "location": "Pune"})
````

````python
# Set default (adds if not present)
student.setdefault("mentor", "Prof. Kumar")
````
