# Python Loops

Loops are used to repeat a block of code multiple times. Python provides two main types of loops: **for loops** and **while loops**. Additionally, loop execution can be controlled using **break**, **continue**, and **pass** statements.

---

## For Loops

### Understanding the `for` loop syntax
- A `for` loop is used to iterate over a sequence (such as a list, tuple, string, or range).
- It automatically moves through each element in the sequence.
- General form:
  - `for variable in sequence:`  
    → Executes the block of code for each item in the sequence.

### Example Concept – Printing "Hello" 100 times
- By using a range of numbers from 0 to 99, the loop repeats 100 times.
- Each repetition prints the word "Hello".

---

## While Loops

### Understanding the `while` loop syntax
- A `while` loop continues executing as long as a given condition is `True`.
- It is useful when the number of iterations is not known in advance.
- General form:
  - `while condition:`  
    → Executes the block of code until the condition becomes `False`.

### Example Concept – Adding 1 to `x` five times
- Start with `x = 0`.
- Repeat the process of adding 1 to `x` until it has been done five times.
- The loop stops once the counter reaches 5.

---

## Loop Control Statements

### Break
- Immediately terminates the loop.
- Useful when you want to stop execution once a certain condition is met.

### Continue
- Skips the current iteration and moves to the next one.
- Useful when you want to ignore specific cases but continue looping.

### Pass
- Acts as a placeholder.
- Does nothing but allows the loop (or function/class) to be syntactically valid.
- Often used during development when the logic is yet to be implemented.

```python
# 1. Print "Hello" 5 times
for i in range(5):
    print("Hello")
```

```python
# 2. Print numbers from 1 to 10
for i in range(1, 11):
    print(i)
```

```python
# 3. Print even numbers between 2 and 20
for i in range(2, 21, 2):
    print(i)
```

```python
# 4. Print each character in a string
for char in "Python":
    print(char)
```

```python
# 5. Print items from a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

# Control Flow Statements in Python

##  [break](ca://s?q=Explain_break_statement_in_Python)
- Terminates the loop immediately.  
- Control jumps out of the loop as soon as `break` is encountered.  
- Useful when a condition is met and continuing the loop is unnecessary.  

---

##  [continue](ca://s?q=Explain_continue_statement_in_Python)
- Skips the current iteration of the loop.  
- Control moves directly to the next iteration without executing the remaining statements in the current cycle.  
- Useful when certain values or cases should be ignored but the loop itself should keep running.  

---

##  [pass](ca://s?q=Explain_pass_statement_in_Python)
- A null operation — it does nothing.  
- Acts as a placeholder where syntactically some statement is required but no action is needed.  
- Often used in function definitions, class declarations, or conditional blocks during development to avoid errors while leaving the logic incomplete.  

---

Summary:  
- **break** → exits the loop entirely  
- **continue** → skips to the next iteration  
- **pass** → does nothing, placeholder  


```python
for i in range(5):
    if i == 3:
        break
    print(i)
# Output: 0, 1, 2
```

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
# Output: 0, 1, 2, 4
```

```python
for i in range(5):
    if i == 3:
        pass
    print(i)
# Output: 0, 1, 2, 3, 4
```
