# Understanding if, else, and elif in Python

## if statement
- The `if` statement is used to test a condition.
- If the condition evaluates to **True**, the block of code under `if` runs.
- If it evaluates to **False**, Python skips that block.

## else statement
- The `else` statement provides an **alternative path**.
- It runs only when the `if` condition is **False**.
- Think of it as the "fallback" option.

## elif statement
- Short for **"else if"**.
- Used when you want to check **multiple conditions** in sequence.
- Python evaluates conditions **top to bottom**:
  - First `if` is checked.
  - If False, each `elif` is checked in order.
  - If none are True, the final `else` runs (if present).

## Key Points
- Only **one block** among `if` / `elif` / `else` executes.
- Conditions are **mutually exclusive** — once one is True, the rest are skipped.
- Indentation is critical in Python: the code inside each block must be indented consistently.
- Logical operators (`and`, `or`, `not`) can be used to combine conditions.
- Nested `if` statements are possible, but `elif` often makes code cleaner.

## Real-world analogy
Imagine deciding what to wear:
- **if** it’s raining → wear a raincoat.
- **elif** it’s cold → wear a jacket.
- **elif** it’s hot → wear a t-shirt.
- **else** → wear whatever you like.

This structure ensures you make **one clear decision** based on the situation.

# Else

```python
# 1. Election Example
votes = 800
if votes > 1000:
    print("Candidate has enough support to win!")
else:
    print("Candidate needs more votes to win.")
```

```python
# 2. Zepto Example
order_value = 250
if order_value > 300:
    print("You qualify for free delivery on Zepto!")
else:
    print("Delivery charges will apply.")
```

```python
# 3. Blinkit Example
items_in_cart = 3
if items_in_cart >= 5:
    print("Blinkit offers you a bulk discount!")
else:
    print("No discount, add more items to qualify.")
```

```python
# 4. Netflix Example
hours_watched = 6
if hours_watched > 8:
    print("You are a binge-watcher on Netflix!")
else:
    print("Your watching habits are moderate.")
```

```python
# 5. Exam Example
marks = 55
if marks >= 75:
    print("You passed with distinction in the exam!")
else:
    print("You did not achieve distinction.")
```

# Elif

```python
# 1. Election Example
votes = 950
if votes > 1000:
    print("Candidate has strong support to win!")
elif votes > 800:
    print("Candidate is in a close race.")
else:
    print("Candidate needs more campaigning.")
```

```python
# 2. Zepto Example
order_value = 280
if order_value >= 500:
    print("You get free delivery + cashback on Zepto!")
elif order_value >= 300:
    print("You qualify for free delivery on Zepto!")
else:
    print("Delivery charges will apply.")
```

```python
# 3. Blinkit Example
items_in_cart = 7
if items_in_cart >= 10:
    print("Blinkit offers you a mega bulk discount!")
elif items_in_cart >= 5:
    print("Blinkit offers you a bulk discount!")
else:
    print("No discount, add more items to qualify.")
```

```python
# 4. Netflix Example
hours_watched = 9
if hours_watched >= 12:
    print("You are a super binge-watcher on Netflix!")
elif hours_watched >= 8:
    print("You are a binge-watcher on Netflix!")
else:
    print("Your watching habits are moderate.")
```

```python
# 5. Exam Example
marks = 68
if marks >= 90:
    print("Excellent! You scored an A+.")
elif marks >= 75:
    print("Great job! You scored an A.")
elif marks >= 50:
    print("You passed the exam.")
else:
    print("You failed. Better luck next time.")
```
